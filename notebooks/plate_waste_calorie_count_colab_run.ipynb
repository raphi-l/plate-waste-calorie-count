```python
import shutil
from pathlib import Path

drive = Path("/content/drive/MyDrive/plate-intake/")

# checkpoints
(drive / "models").mkdir(exist_ok=True)
for ckpt in Path("models/").glob("*.pt"):
    shutil.copy(ckpt, drive / "models" / ckpt.name)
    print(f"✓ {ckpt.name}")

# mlflow
if Path("mlflow.db").exists():
    shutil.copy("mlflow.db", drive / "mlflow.db")
    print("✓ mlflow.db")

print("All done — safe to close")
```
